```python
import pandas as pd
```

```python
df = pd.read_csv('sales_with_nans.csv')

df.head(20)
```

```python
#df['Brand'] = df['Brand].fillna(Unknown)

# replace missing values


df['Group'] = df['Group'].fillna('Unknown')
df['Customer_Segment'] = df['Customer_Segment'].fillna('Unknown')
df['Sales_Before'] = df['Sales_Before'].fillna(df['Sales_Before'].mean())
df['Sales_After'] = df['Sales_After'].fillna(df['Sales_After'].mean())
df['Customer_Satisfaction_Before'] = df['Customer_Satisfaction_Before'].fillna(df['Customer_Satisfaction_Before'].mean())
df['Customer_Satisfaction_After'] = df['Customer_Satisfaction_After'].fillna(df['Customer_Satisfaction_After'].mean())
df['Purchase_Made'] = df['Purchase_Made'].fillna('Unknown')

df.head(20)
```

```python
# total sales before

total_sales_before = df['Sales_Before'].sum()

print(total_sales_before)

total_sales_after = df['Sales_After'].sum()

print(total_sales_after)
```

```python
# average sales before & after

avg_before = df['Sales_Before'].mean()

avg_after = df['Sales_After'].mean()

print(avg_before)
print(avg_after)
```

```python
# satisfaction before & after

satis_before = df['Customer_Satisfaction_Before'].mean()

satis_after = df['Customer_Satisfaction_After'].mean()

print(satis_before)
print(satis_after)
```

```python
# min and max sales

min_sales_before = df['Sales_Before'].min()
max_sales_before = df['Sales_Before'].max()

min_sales_after = df['Sales_After'].min()
max_sales_after = df['Sales_After'].max()

print(min_sales_before)
print(max_sales_before)

print(min_sales_after)
print(max_sales_after)
```

```python
# calculate change in avg sales and avg satisfaction before and after

avg_sales_change = df['Sales_After'].mean() - df['Sales_Before'].mean()

print(avg_sales_change)
```

```python
avg_satis_change = df['Customer_Satisfaction_After'].mean() - df['Customer_Satisfaction_Before'].mean()

print(avg_satis_change)
```

```python
# filtering & subsets

# Get all rows for a specific customer_segment e.g high value

high_val = df[df['Customer_Segment'] == 'High Value']

print(high_val.head(10))
```
