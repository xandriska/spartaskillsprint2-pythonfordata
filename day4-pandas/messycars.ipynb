```python
import pandas as pd
```

```python
df = pd.read_csv('messy_cars.csv')
df.head(20)
```

```python
# Filling in missing values

df['Brand_Filled'] = df['Brand'].fillna('Unknown_Brand')
df['Money_Filled'] = df['Money'].fillna(df['Money'].mean())
df['Miles_Filled'] = df['Miles'].fillna(df['Miles'].mean())
df['Colour_Filled'] = df['Color'].fillna('Unknown_Colour')
df.head(20)


## NOTE: for a cleaner version of this without creating new columns for the missing values, you can do df['Brand'] = df['Brand].fillna(Unknown)
## and it will simply replace those fields of NaN with Unknown in the pre-existing columns.
```

```python
# filtering

# older_than_30 = df[df['Age'] > 17]

# older_than_30.head(20)

high_miles_brands = df[df['Miles_Filled'] > 100000].sort_values(by='Miles_Filled', ascending=False)

high_miles_brands.head(20)
```

```python
money_desc = df.sort_values(by='Money_Filled', ascending=False)
money_desc.head(20)
```

```python
df.columns = ['Car_ID', 'Brand', 'Model', 'Year', 'Price', 'Mileage', 'Colour', 'Brand_Filled', 'Money_Filled', 'Miles_Filled', 'Colour_Filled']

df.head(20)
```

```python
# GROUP BY

# group by brand

brand_totals = df.groupby('Brand')['Miles_Filled'].mean().reset_index(name='Average Mileage')
brand_totals.head(20)
```

```python
# multiple groupbys - aggregation

gp_mod_colour = df.groupby(['Model', 'Colour']).agg(
    avg_prices=('Price', 'mean'),
    total_cars=('Car_ID', 'count')
).reset_index()

gp_mod_colour.head(20)

```

```python
gp_by_colour = df.groupby('Colour')['Miles_Filled'].mean().reset_index(name='Average Mileage')

gp_by_colour.head(20)
```

```python
gp_by_brand_maxp = df.groupby('Brand')['Price'].max().reset_index(name='Max Price')

gp_by_brand_maxp.head(20)
```

```python
#Group by brand and color and count how many cars are in each combination

brand_colour_count = df.groupby(['Brand', 'Colour']).agg(
    car_count=('Car_ID', 'count')
).reset_index()

brand_colour_count.head(20)
```

```python
#Group by brand and calculate both the average mileage and total money spent in one .agg()

brand_price_avg = df.groupby(['Brand']).agg(
    avg_mileage=('Mileage', 'mean'),
    total_cost=('Price', 'sum')
).reset_index()

brand_price_avg.head(20)
```
