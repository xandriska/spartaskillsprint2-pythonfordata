```python
import pandas as pd
```

```python
df = pd.read_csv('students_data_messy.csv')

df.head()
```

```python
# drop rows if val missing

df_dropped = df.dropna()

df_dropped.head()
```

```python
# Filling in missing values
df.head(20)
df["Age_Filled"] = df["Age"].fillna(0)
df.head(20)
```

```python
df['score_filled'] = df['Score'].fillna(df['Score'].mean())

df.head(20)
```

```python
# filtering

older_than_30 = df[df['Age'] > 17]

older_than_30.head(20)
```

```python
# age and score (multiple conditions)

high_score = df[(df['Age'] > 17) & (df['Score'] > 85)]

high_score.head(20)
```

```python
# renaming cols

df_renamed = df.rename(columns={'Age': 'Age_Years', 'Score': 'Test_Score'})

df_renamed.head(20)
```
