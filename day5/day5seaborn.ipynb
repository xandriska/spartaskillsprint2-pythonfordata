```python
tips = sns.load_dataset('tips')

tips
```

```python
totalbill_tip = sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size='size', palette='YlGnBu')
```

```python
tip_size = sns.scatterplot(x='tip', y='size', data=tips)
```

```python
total_gender = sns.scatterplot(x='total_bill', y='sex', data=tips, hue='sex')
```

```python
total_day = sns.scatterplot(x='time', y='total_bill', data=tips, hue='day')
```

```python
sns.histplot(tips['tip'], kde=True, bins=20)
```

```python
sns.barplot(x='sex', y='tip', data=tips, hue='sex', palette='YlGnBu')
```

```python
sns.boxplot(x='day', y='tip', data=tips, hue='sex')
```

```python
sns.stripplot(x='day', y='tip', data=tips, hue='sex', dodge=True)
```

```python
sns.jointplot(x='tip', y='total_bill', data=tips, kind='reg')
```

```python
# pairplot - produces scatterplots or histograms for every possible combination
# heatmap
#explore these with docs

```

```python
sns.pairplot(data=tips)
```

```python
sns.pairplot(data=tips, x_vars=['tip', 'total_bill'], y_vars=['day', 'time'], kind='hist')
```
